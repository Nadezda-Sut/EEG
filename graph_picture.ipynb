```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D']
G.add_nodes_from(nodes)

# Add edges
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D')]
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

# Nodes
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')

# Edges
nx.draw_networkx_edges(G, pos, edgelist=edges, width=2)

# Labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.title("Graph Network with Four Nodes")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Define modules
modules = {
    'Module1': ['A1', 'A2', 'A3', 'A4'],
    'Module2': ['B1', 'B2', 'B3', 'B4'],
    'Module3': ['C1', 'C2', 'C3', 'C4', 'C5'],
    'Module4': ['D1', 'D2', 'D3', 'D4', 'D5', 'D6']
}

# Define edges within each module
edges_within = {
    'Module1': [('A1', 'A2'), ('A2', 'A3'), ('A3', 'A4'), ('A4', 'A1')],
    'Module2': [('B1', 'B2'), ('B2', 'B3'), ('B3', 'B4'), ('B4', 'B1')],
    'Module3': [('C1', 'C2'), ('C2', 'C3'), ('C3', 'C4'), ('C4', 'C5'), ('C5', 'C1')],
    'Module4': [('D1', 'D2'), ('D2', 'D3'), ('D3', 'D4'), ('D4', 'D5'), ('D5', 'D6'), ('D6', 'D1')]
}

# Add nodes and edges for each module
for module, nodes in modules.items():
    G.add_nodes_from(nodes)
    G.add_edges_from(edges_within[module])

# Define reach club connections
reach_club_connections = [
    ('A1', 'B1'),  # Connection between Module1 and Module2
    ('A2', 'C1'),  # Connection between Module1 and Module3
    ('B3', 'D1'),  # Connection between Module2 and Module4
    ('C4', 'D4')   # Connection between Module3 and Module4
]

# Add reach club edges
G.add_edges_from(reach_club_connections)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

# Nodes
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')

# Edges
nx.draw_networkx_edges(G, pos, width=2)

# Labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

plt.title("Graph Network with Four Modules Connected via Reach Club")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Define modules
modules = {
    'Module1': ['A1', 'A2', 'A3', 'A4'],
    'Module2': ['B1', 'B2', 'B3', 'B4'],
    'Module3': ['C1', 'C2', 'C3', 'C4', 'C5'],
    'Module4': ['D1', 'D2', 'D3', 'D4', 'D5', 'D6']
}

# Add nodes and fully connect each module (complete graph)
for module, nodes in modules.items():
    G.add_nodes_from(nodes)
    for i in range(len(nodes)):
        for j in range(i + 1, len(nodes)):
            G.add_edge(nodes[i], nodes[j])

# Define the central hub
hub = 'Hub'

# Add the hub node
G.add_node(hub)

# Define connections between modules and the hub
hub_connections = [
    (hub, 'A1'),  # Hub to Module 1
    (hub, 'B1'),  # Hub to Module 2
    (hub, 'C1'),  # Hub to Module 3
    (hub, 'D1')   # Hub to Module 4
]

# Add hub edges
G.add_edges_from(hub_connections)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

# Nodes
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
nx.draw_networkx_nodes(G, pos, nodelist=[hub], node_size=700, node_color='red')  # Different color for hub

# Edges
nx.draw_networkx_edges(G, pos, width=2)

# Labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

plt.title("Graph Network with Fully Connected Modules and Central Hub")
plt.show()
```
